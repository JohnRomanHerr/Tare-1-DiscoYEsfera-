```python
import matplotlib.pyplot as plt
import numpy as np
rad = 15
num = 1000

theta = np.random.uniform(0.0, 2.0*np.pi, num)
r = rad * np.random.uniform(0.0, 1.0, num)
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y, "ro", ms=1)
plt.axis([-25, 25, -25, 25])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
rad = 15
num = 1000

theta = np.random.uniform(0.0, 2.0*np.pi, num)
r = rad * np.sqrt(np.random.uniform(0.0, 1.0, num))
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y, "ro", ms=1)
plt.axis([-25, 25, -25, 25])
plt.show()
```
